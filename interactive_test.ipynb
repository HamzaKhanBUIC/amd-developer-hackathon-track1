# AMD Hackathon - Track 1 - Interactive Test Notebook
This notebook tests the `process_tasks_batch` function to verify that our zero-token routing perfectly divides tasks between the local CPU model and the Fireworks API.

```python
import sys
import os
import asyncio

# Add backend to path so we can import the agent
sys.path.append(os.path.join(os.getcwd(), 'backend'))
from agent import process_tasks_batch
```

```python
# Define 4 test tasks spanning different categories
tasks = [
    {"id": "1", "prompt": "Is this positive or negative? I absolutely love this new AMD CPU!"},
    {"id": "2", "prompt": "Extract entities from: Google and Microsoft announced new AI models in Seattle today."},
    {"id": "3", "prompt": "Calculate exactly: What is 15 * 24 + 100?"},
    {"id": "4", "prompt": "Write a python function to recursively reverse a linked list."}
]

# Execute tasks asynchronously (Jupyter already has an event loop running, so we await directly)
results = await process_tasks_batch(tasks)
```

```python
import json

for r in results:
    print("-" * 50)
    print(f"Task ID: {r['task_id']}")
    print(f"Routed To: {r['routing']['model']} (Layer: {r['routing']['layer']})")
    print(f"Answer: {r['answer']}\n")
```
